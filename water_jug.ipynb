```python
dp = []
visited = []

amax = 4
bmax = 3

areq = 2
breq = 0

curr_a = 0
curr_b = 0

count = 0

for i in range(amax+1):
    lst=[]
    for j in range(bmax+1):
        lst.append(-1)
    dp.append(lst)
    

for i in range(amax+1):
    lst=[]
    for j in range(bmax+1):
        lst.append(-1)
    visited.append(lst)
    
track = []
maintrack = []
# print(dp)
# print(visited)


# options : afull, bfull, aempty, bempty, atransfer, btransfer

def helper(a, b):
    
    if(a<0 or b<0 or a>amax or b>bmax):
        return -1

    if(a==areq and b==breq):
        maintrack.append(track)
        dp[a][b] = 1
        return 1
    
#     if(a==amax and b==bmax):
#         dp[a][b] = 0
#         return 0
    
    
    if(dp[a][b] != -1):
        return dp[a][b]
    
    if(visited[a][b]==1):
        return 0
    
    visited[a][b] = 1
    dp[a][b] = 9999
    print(a,b)
    
    x = (amax,b)
    track.append(x)
    
    #case 1 : fill a 
    val = helper(amax, b)
    dp[amax][b] = val
    if(val>0):
        dp[a][b] = min(dp[a][b], val+1)
#         return val
        
    track.pop()
    x = (a,bmax)
    track.append(x)
    
    #case 2 : fill b
    val = helper(a, bmax)
    dp[a][bmax] = val
    if(val>0):
        dp[a][b] = min(dp[a][b], val+1)
#         return val
    
    
    track.pop()
    x = (0,b)
    track.append(x)
        
    #case 3 : empty a 
    val = helper(0, b)
    dp[0][b] = val
    if(val>0):
        dp[a][b] = min(dp[a][b], val+1)
#         return val
        
        
    track.pop()
    x = (a,0)
    track.append(x)
    
    #case 4 : empty b
    val = helper(a, 0)
    dp[a][0] = val
    if(val>0):
        dp[a][b] = min(dp[a][b], val+1)
#         return val
    
    
    
    
    #case 5: transfer from a to b
    tempb = min(bmax, a+b)
    added_in_b = tempb - b
    rem_in_a = a - added_in_b;
    
    track.pop()
    x = (rem_in_a,tempb)
    track.append(x)
    
    val = helper(rem_in_a, tempb)
    dp[rem_in_a][tempb] = val
    if(val>0):
        dp[a][b] = min(dp[a][b], val+1)
#         return val
    

    
    
    #case 6: transfer from b to a
    tempa = min(amax, a+b)
    added_in_a = tempa - a
    rem_in_b = b - added_in_a
    
    track.pop()
    x = (tempa,rem_in_b)
    track.append(x)
    
    val = helper(tempa, rem_in_b)
    dp[tempa][rem_in_b] = val
    if(val>0):
        dp[a][b] = min(dp[a][b], val+1)
#         return val
    
    
    track.pop()

    
    dp[a][b] = -1  
    return -1
    

    
print(helper(0,0))  

print(*dp)

print(track)
```
